```python
import requests                                     #для получения запросов
from bs4 import BeautifulSoup                       #для раскодирования html
```

```python
base_url = 'https://kinoteatr.ru/'
```

```python
r = requests.get(f'{base_url}/raspisanie-kinoteatrov/')
```

```python
r.text
```

```python
soap = BeautifulSoup(r.text, 'lxml')
```

```python
cinemas = []
for i in soap.findAll('div', class_= 'col-md-12 cinema_card'):
    name = i.find('h3', class_='title movie_card_title').text.strip()
    href = i.find('a')['href'].strip()
    address = i.findAll('span', class_='sub_title')[0].text.strip()
    cinemas.append({
        'name': name,
        'href': href,
        'address': address  
    })
#     print(name, href, address, sep='\n', end='\n------------------------\n')
```

```python
import time
import pprint
```

```python
dict_rezult={}
for cinema in cinemas:
    dict_today_tomorrow = {}
    r_today = requests.get(cinema['href'])
    soap_today = BeautifulSoup(r_today.text, 'lxml')
    dict_timetable_today = {} 
    films_today = soap_today.findAll('div', class_='shedule_movie bordered gtm_movie')
    for film_today in films_today:
        film_info_today = {}
        film_schedule_today = film_today.find('div', class_='shedule_movie_sessions col col-md-8')
        sessions_today = film_schedule_today.findAll('a', class_='shedule_session')
        this_film_sessions_today = {}
        for ind, session_today in enumerate(sessions_today,1):
            this_film_sessions_today[ind] = session_today.find('span', class_='shedule_session_time').text.strip()
        film_info_today = {
            'format': film_today['data-gtm-format'],
            'genre': film_today['data-gtm-list-item-genre'],
            'raiting_sub': film_today.findAll('i', class_='raiting_sub')[0].text.strip(),
            'timetable': this_film_sessions_today
        }
        dict_timetable_today[film_today['data-gtm-list-item-filmname']] = film_info_today
        
 
    r_tomorrow = requests.get(cinema['href']+'?day=tomorrow')    
    soap_tomorrow = BeautifulSoup(r_tomorrow.text, 'lxml')
    dict_timetable_tomorrow = {} 
    films_tomorrow = soap_tomorrow.findAll('div', class_='shedule_movie bordered gtm_movie')
    for film_tomorrow in films_tomorrow:
        film_info_tomorrow = {}
        film_schedule_tomorrow = film_tomorrow.find('div', class_='shedule_movie_sessions col col-md-8')
        sessions_tomorrow = film_schedule_tomorrow.findAll('a', class_='shedule_session')
        this_film_sessions_tomorrow = {}
        for ind, session_tomorrow in enumerate(sessions_tomorrow,1):
            this_film_sessions_tomorrow[ind] = session_tomorrow.find('span', class_='shedule_session_time').text.strip()
        dict_timetable_tomorrow[film_tomorrow['data-gtm-list-item-filmname']] = this_film_sessions_tomorrow
        film_info_tomorrow = {
            'format': film_tomorrow['data-gtm-format'],
            'genre': film_tomorrow['data-gtm-list-item-genre'],
            'raiting_sub': film_tomorrow.findAll('i', class_='raiting_sub')[0].text.strip(),
            'timetable': this_film_sessions_tomorrow
        }
        dict_timetable_tomorrow[film_tomorrow['data-gtm-list-item-filmname']] = film_info_tomorrow
    
    dict_today_tomorrow['TODAY'] = dict_timetable_today
    dict_today_tomorrow['TOMORROW'] = dict_timetable_tomorrow
    dict_rezult[cinema['name']] = dict_today_tomorrow
    
pprint.pprint(dict_rezult)
```
