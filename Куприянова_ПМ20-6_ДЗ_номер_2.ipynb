```python
def typical_student_this_group(id_group):
    """
    Нахождение типичного студента из данной группы
    t_s_t_g - сокращение для словаря, в котором зранится один случайный студент из данной группы
    """
    t_s_t_g = {}
    for i in lst_students:
        if i['group_id'] == id_group:
            t_s_t_g = i
            break
    return(t_s_t_g)
```

```python
def check_teacher_group(id_teach, t_s_t_g):
    """
    Проверка на то, что данный учитель преподает у данной группы
    """
    boolean = False 
    for i in lst_results:
        if i['student_id'] == t_s_t_g['id'] and i['teacher_id'] == id_teach:
            boolean = True
    return(boolean)
```

```python
def check_student(student_id, id_group):
    """
    Проврка, что данный студент входит в данную группу
    """
    boolean = False
    for k in lst_students:
        if k['id'] == student_id and k['group_id'] == id_group: 
            boolean = True
    return(boolean)
```

```python
def results_of_this_group(id_group):
    """
    Подсчет колличества студентов из ДАННОЙ группы, набравших 2,3,4,5
    """
    dict_results = {} 
    for i in lst_subjects:                  
        dict_results[i['subject_name']] = {}
    
    for i in lst_subjects:
        count2 = 0
        count3 = 0
        count4 = 0
        count5 = 0
        for j in lst_results:
            if i['id'] == j['subject'] and check_student(j['student_id'], id_group):
                if 86<= int(j['att1']) + int(j['att2']) + int(j['exam']):
                    count5 +=1
                if 70<= int(j['att1']) + int(j['att2']) + int(j['exam']) <=85:
                    count4 +=1
                if 50<= int(j['att1']) + int(j['att2']) + int(j['exam']) <=69:
                    count3 +=1
                if 50> int(j['att1']) + int(j['att2']) + int(j['exam']):
                    count2 +=1
        dict_results[i['subject_name']] = {'5': count5, '4': count4, '3': count3, '2': count2}
    
    return(dict_results)
        
```

```python
def first_function(id_teach, id_group):
    """
    Функция возвращает словарь, в котором ключами являются наименования предметов, а значениями словари, 
    которые хранят в себе информации о количестве студентов, сдавших на 5, 4, 3 и 2 по данному предмету.
    """
    count = 0
    for i in lst_teachers:
        if id_teach in i.values():
            count +=1 
    if count == 0:
        return(print('None'))
    else:
        typ_student_this_group = typical_student_this_group(id_group)
        if check_teacher_group(id_teach,typ_student_this_group): 
            return(results_of_this_group(id_group))
        else:
            return(print('False'))
       
```

```python
def second_function(teacher, to_json = False):
    """
    Функция возвращает словарь- ключами являются наименования групп, в которых он преподавал,
    а значениям - результат выполнения первой функции.
    """
    
    lst_name_teacher = teacher.split(' ')
    for i in lst_teachers:
        if i['last_name'] == lst_name_teacher[0]:
            teacher_id = i['id']
    
    dict_res = {}
    for i in lst_groups:
        typ_student_this_group = typical_student_this_group(i['id'])
        if check_teacher_group(teacher_id,typ_student_this_group): 
            dict_res[i['id']] = first_function(teacher_id,i['id'])
            
    if too_json == '':
        return(dict_res)
    else:
        to_json = too_json
        with open(to_json, 'w', encoding = "utf-8") as file:
            json.dump(dict_res, file, ensure_ascii=False)
        return('True')
            
```

```python
def inner_dict(group_id, subject_id):
    """
    Функция должна возвращать словарь со статистикой по группам по данному предмету. 
    Ключами словаря являются id группы, а значениями словари.
    Внутренний словарь имеет ключи group_name (наименование группы), stats (статистика). 
    Значением ключа статистика является словарь, который возвращает количество студентов получившие оценку 5, 
    количество студентов получившие оценку 4, количество студентов получившие оценку 3, 
    количество студентов получившие оценку 2. 
    """
    group_name = ''
    for i in lst_groups:
        if i['id'] == group_id:
            group_name = i['name']
    
    stats = {}
    dict_rezult = {}
    count2 = 0
    count3 = 0
    count4 = 0
    count5 = 0
    for j in lst_results:
        if check_student(j['student_id'], group_id) and j['subject'] == subject_id:
            if 86<= int(j['att1']) + int(j['att2']) + int(j['exam']):
                count5 +=1
            if 70<= int(j['att1']) + int(j['att2']) + int(j['exam']) <=85:
                 count4 +=1
            if 50<= int(j['att1']) + int(j['att2']) + int(j['exam']) <=69:
                count3 +=1
            if 50> int(j['att1']) + int(j['att2']) + int(j['exam']):
                count2 +=1
    stats = {'5': count5, '4': count4, '3': count3, '2': count2} 
    dict_rezult[group_name] = stats
    return(dict_rezult)
```

```python
def third_function(year_entry, subject_name, to_json = False):
    """
    Главная функция третьей задачи
    """
    d_res = {}
    for i in lst_groups:
        if i['entry_year'] == year_entry:
            d_res[i['id']] = {}
    
    id_subject = ''
    for j in lst_subjects:
        if j['subject_name'] == subject_name:
            id_subject = j['id']
    
    for gr_id in d_res:
        d_res[gr_id] = inner_dict(gr_id, id_subject)
    
    if too_json == '':
        return(d_res)
    else:
        to_json = too_json
        with open(to_json, 'w', encoding = "utf-8") as file:
            json.dump(d_res, file, ensure_ascii=False)
        return('True')
```

```python
def results_themselves(st_id, sub_id):
    """
    Получение списка результатов данного студента по данному предмету
    """
    results = {}
    for i in lst_results:
        if i['student_id'] == st_id and i['subject'] == sub_id:
            results['att1'] = i['att1']
            results['att2'] = i['att2']
            results['exam'] = i['exam']
            results['total'] = i['total']
    return(results)
            
```

```python
def first_funct(student_id):
    """
    Функция возвращает сумму баллов студента по каждому из контрольных мероприятий, в том числе total
    """
    dict_results = {}
    for i in lst_results:
        if i['student_id'] == student_id:
            for j in lst_subjects:
                if i['subject'] == j['id']:
                    subject_id = j['id']
                    dict_results[j['subject_name']] = results_themselves(student_id, subject_id)
    return(dict_results)
```

```python
def sum_test(test, stud_id):
    """
    Функция возвращает сумму всех элесентов test из результатов по всем предметам
    """
    sum_test = 0
    for i in lst_results:
        if i['student_id'] == stud_id:
            sum_test += int(i[test])
    return(sum_test)  
```

```python
def student_info(stud_id):
    """
    Получение списка, содержащего всю информацию о студенте из student.csv
    """
    info = {}
    for i in lst_students:
        if i['id'] == stud_id:
            info = i
    return(info)
```
